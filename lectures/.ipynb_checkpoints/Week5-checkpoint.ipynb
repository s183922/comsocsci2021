Lexical Diversity is a measure of how many different word at text uses. A measure of 'richness'.

```python
n = 75
lowercase_no_punctuation = [w.lower() for w in text2 if w.isalpha()]
fdist = nltk.FreqDist(lowercase_no_punctuation)
most_common = fdist.most_common(n)
words, counts = zip(*most_common)
fig, ax = plt.subplots(2,dpi = 200)
ax[0].plot(np.array(counts)/sum(counts), alpha = .7)
ax[1].plot(np.cumsum(counts)/sum(counts), alpha = .7,  c = 'orange')

ax[0].set_xticks(np.arange(n))
ax[0].set_xticklabels(words, rotation = 90, fontsize = 5)
ax[0].grid(linestyle = '--')
ax[0].set_ylabel("Freq. top 75  words", fontsize = 6)

ax[1].set_xticks(np.arange(n))
ax[1].set_xticklabels(words, rotation = 90, fontsize = 5)
ax[1].grid(linestyle = '--')
ax[1].set_ylabel("Cum. freq. top 75 words", fontsize = 6)
plt.show()

```
